# PLOTLY AND CUFFLINKS FOR INTERACTIVE DATA VISUALIZATION

```python
import numpy as np
import pandas as pd
```

```python
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly import __version__
print(__version__)
```

```python
# Connect notebook to Javascript
init_notebook_mode(connected=True)
```

```python
# Allow visualization to run offline
cf.go_offline()
```

```python
# Data
df = pd.DataFrame(np.random.randn(100, 4), columns='A B C D'.split())
df.head()
```

```python
df2 = pd.DataFrame({'Category': ['a', 'b', 'c'], 'Values': [32,43,50]})
df2.head()
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame(
    {"x": [1, 2, 3, 4, 5], "y": [10, 20, 30, 20, 10], "z": [5,4,3,2,1]}
)
```

```python
# 3D Surface Plot
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
df.iplot(kind='hist')
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C', colorscale='rdylbu')
```
