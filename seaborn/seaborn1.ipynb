# Seaborn Contined

### Matrix Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
# heatmaps: Matrix plots need you data to be in a matrix form (where you can matchup column data to the rows) for it to be able to work
fp = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(fp,cmap='magma')
```

### Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# PairGrid gives you more control than pairplot
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# Facet Grid
grid = sns.FacetGrid(data=tips, col='time', row='smoker')
grid.map(sns.histplot, 'total_bill')
```

```python
grid = sns.FacetGrid(data=tips, col="time", row="smoker")
grid.map(plt.scatter, "total_bill", "tip")
```

### Regression Plots

```python
# Seperating by hue
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s': 20})
```

```python
# Seperating by grids
sns.lmplot(
    x="total_bill",
    y="tip",
    data=tips,
    col="day",
    hue='sex',
    aspect=0.6,
    
)
```

### Styling plots

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,3))
sns.set_context('notebook')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.lmplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="sex",
    palette='seismic'
)
# Color palettes available at matplotlib colormap documentation
```
