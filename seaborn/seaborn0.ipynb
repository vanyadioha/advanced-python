# Python for data visualization: Seaborn

### Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
# For visualizing Univariate data. distplot() is deprecated so histplot and displot should be used
sns.histplot(tips['total_bill'],kde=True, bins=30)
```

### Joint plot

```python
# Joint plot is used for visualizing bivariate data
sns.jointplot(data=tips, x='total_bill', y='tip',kind='scatter')
# There are multiple kinds such as scatter, hex, kde, reg(regression etc)
```

### Pair plots

```python
# Plots pair wise relationships across an entire dataframe. Hence, one just needs to pass in the dataframe variable and sns generates alll the pairs
sns.pairplot(tips, hue='day',palette='coolwarm')
# Hue colors the data points bases off of categorical columns in the data frame. like sex, time of day, seasons etc
```

### Rug Plot

```python
# Another plot for visualizing univariate data
sns.rugplot(tips['total_bill'])
```

### Categorical plots

Categorical plots are used to visualize the relationship between a categorical column such as gender, weather, time of day etc and a numerical column or another categorical column

```python
# Bar plot (On default this shows the mean of both categories. The aggregation can be set to another statistical iperation or a custom function)
import numpy as np
sns.barplot(data=tips, x='day', y='total_bill', estimator=np.var, hue='day', palette='coolwarm')
```

```python
# Count plot is used to count the number of occurences of a variable in a dataset
sns.countplot(data=tips, x='day')
```

```python
# Box plot shows the distribution of quatitative data in a way that facilitates comparison b/w variables
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
# Violin plots
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# Striplot
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# Swarm plot
sns.violinplot(x="day", y="total_bill", data=tips, hue="day")
sns.swarmplot(x="day", y="total_bill", data=tips, hue="day", palette='dark:black')
```
