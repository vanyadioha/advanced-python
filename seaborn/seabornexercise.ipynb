```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
sns.set_style('whitegrid')
t = sns.load_dataset('titanic')
t.head()
```

```python
sns.jointplot(x='fare', y='age', data=t)
```

```python
sns.histplot(t['fare'], kde=True, color='r', bins=60)
```

```python
sns.boxplot(x='class', y='age', data=t, hue='class', palette='magma')
```

```python
sns.swarmplot(x='class', y='age', data=t, hue='class', palette='seismic')
```
