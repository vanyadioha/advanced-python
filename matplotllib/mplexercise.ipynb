# MatPlotLib Exercise

```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
# Setting MPL to display inline
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('Abscissa')
ax.set_ylabel('Ordinate')
ax.set_title('Sample Plot')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
axes[0].plot(x,y,color='red', lw=2, ls='-.')
axes[1].plot(x,z, marker='o', alpha=0.5)
```
