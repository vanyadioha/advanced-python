# Python for data visualization: MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

### Functional Method of creating plots

```python
plt.plot(x,y)
# plt.show() if You are not in the jupyter nb
```

```python
# Plotting with a list
a = [1,2,3,4,5,6]
b = []
for i in a:
    b.append(i * 2)
plt.plot(a,b)
```

```python
# Plotting from a series
c = pd.Series([1,2,3,4,5])
d = c - 1
plt.plot(c,d, 'r-')
plt.xlabel('Ordinate')
plt.ylabel('Abscissa')
plt.title('Simple Line Plot')
# Color and line styles can also be added
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(a,b,'g')
plt.subplot(2,2,3)
plt.plot(c,d,'b')
```

### OOP method of using MatPlotLib

```python
plot = plt.figure()
axes = plot.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
# Creating Insert plots
plot1 = plt.figure()
axes1 = plot1.add_axes([0.1,0.1,0.8,0.8])
axes2 = plot1.add_axes([0.2, 0.5, 0.4,0.3])
axes1.plot(x,y,'r')
axes2.plot(y,x, 'g')
axes1.set_title('LARGE')
axes2.set_title('SMALL')
```

```python
# Creating Sub plots
fig,coords = plt.subplots(nrows=1, ncols=2)
# for coord in coords:
#     coord.plot(x,y)
coords[0].plot(x,y)
coords[1].plot(y,x)
# Introducing spacing to take care of overlaps
plt.tight_layout()
```

```python
# FigSize and legend
# VSCODE doesn't have the capability to show plots with edited fig sizes
chart = plt.figure()
axes = chart.add_axes([0.1,0.1,0.8,0.5])
axes.plot(x,y,label='Y=X**2')
axes.plot(x,y*x,label='Y=X**3')
axes.set_xlabel('Abscissa')
axes.set_ylabel('Ordinate')
axes.set_title('Parabolic Plot')
axes.legend()
```

```python
# Saving a plot
chart.savefig('mychart.png')
```
